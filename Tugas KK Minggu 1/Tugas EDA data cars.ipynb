```python
data = data.drop(['Engine Fuel Type', 'Engine Cylinders', 'Market Category', 'Vehicle Style', 'Number of Doors', 'Vehicle Size'], axis=1)
data.head(5) # mendrop data yang tidak terlalu mempengaruhi tingkatan data cars
```

```python
duplicate_rows_data = data[data.duplicated()] # mengecek data yang bernilai sama pada baris
```

```python
data = data.drop_duplicates() # melakukan drop pada data yang sama pada data cars
data.head(5)
```

```python
print(data.isnull().sum()) # setelah mendrop data yang sama, selanjutnya memeriksa nilai null pada data cars
```

```python
data = data.dropna()    # mendrop nilai null
data.count()            # mengecek data cars, apakah semua data null sudah di drop atau belum
```
